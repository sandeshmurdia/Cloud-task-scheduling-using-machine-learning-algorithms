```python
from mealpy.swarm_based.PSO import BasePSO
import json 
import time
from mealpy.utils.visualize import *
import numpy as np
```

```python
def Fun(solution):
  list1 = [0] * data['vmcount'][0]
  val = 0.0
  #print(solution)
  #print(int(round(solution[0])))
  for i in range(solution.size):
    solution[i]=round(solution[i])
    if(round(solution[i])!=0.0):
      list1[int(round(solution[i]))-1]=list1[int(round(solution[i]))-1] + data['task'][0]/data['VM'][int(round(solution[i]))-1] 
    else:
      list1[int(round(solution[i]))]=list1[int(round(solution[i]))-1] + data['task'][0]/data['VM'][int(round(solution[i]))] 
  #print(solution)
  return max(list1) 
    
start_time = time.time()
```

```python
vm=5
tcount=10000
number=500
f = open('data_'+str(tcount)+"_"+str(vm)+'.json') 
data = json.load(f)  
size=number
vmcount= data['vmcount'][0]

problem_dict1 = {
    "fit_func": Fun,
    "lb": [0 for i in range(size)] ,
    "ub": [vmcount for i in range(size)],
    "minmax": "min",
    "verbose": True,
}
model1 = BasePSO(problem_dict1, epoch=100, pop_size=100, pr=0.03)
model1.solve()
```

```python
model1.history.save_global_objectives_chart(filename="PSO_5_500/goc")
model1.history.save_local_objectives_chart(filename="PSO_5_500/loc")
model1.history.save_runtime_chart(filename="PSO_5_500/rtc")
model1.history.save_diversity_chart(filename="PSO_5_500/dc")
```
